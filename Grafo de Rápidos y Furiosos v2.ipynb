# Grafo de Rápidos y Furiosos

La saga de rápidos y furiosos es una de las franquicias más exitosas y populares en la historia del cine. Comenzó en 2001 con el lanzamiento de "The Fast and the Furious" y ha crecido hasta incluir múltiples secuelas, spin-offs, y una expansión en diferentes medios.
### Origen
La franquicia comenzó como una serie de películas centradas en carreras callejeras ilegales y la cultura automovilística de Los Ángeles. La primera película presenta a Brian O'Conner (Paul Walker), un policía encubierto que se infiltra en una banda de corredores callejeros liderada por Dominic Toretto (Vin Diesel). A lo largo de la película, Brian debe elegir entre cumplir con su deber como policía o su lealtad a su nuevo grupo de amigos.
### Evolución
Con el paso del tiempo, la saga evolucionó de su enfoque inicial en carreras callejeras a incluir elementos de acción, atracos, y espionaje internacional. A medida que la trama se expandía, también lo hacía su elenco, introduciendo personajes interpretados por actores como Dwayne "The Rock" Johnson, Jason Statham, y Charlize Theron, entre otros.
### Temas y Características
**Familia y Lealtad**: Uno de los temas centrales de la saga es la importancia de la familia y la lealtad. Dominic Toretto y su equipo consideran a su grupo como una familia, unida no por lazos de sangre sino por su apoyo y confianza mutua.

**Acción y Efectos Especiales**: Las películas son conocidas por sus espectaculares secuencias de acción, que incluyen persecuciones de autos a alta velocidad, acrobacias imposibles, y explosiones impresionantes.

**Diversidad Cultural**: La saga ha sido elogiada por su diverso elenco y su representación de diferentes culturas, lo cual ha contribuido a su amplio atractivo global.

## ------------------------------------------Grafo---------------------------------------

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
extended_characters = [
    "Dominic Toretto", "Brian O'Conner", "Mia Toretto", "Letty Ortiz", 
    "Roman Pearce", "Tej Parker", "Han Lue", "Gisele Yashar", 
    "Luke Hobbs", "Deckard Shaw", "Jakob Toretto", "Elena Neves", 
    "Mr. Nobody", "Ramsey", "Magdalene Shaw", "Hattie Shaw", "Owen Shaw", 
    "Sean Boswell", "Neela", "Twinkie", "Suki", "Agente Bilkins", "Carter Verone", "Monica", "Cypher", "Vince", "Jesse", "Leon", "Braga", "Fenix", "Takeshi", "Reyes", "Dante", "Johnny Tran"
]
```

```python
G.add_nodes_from(extended_characters)
```

```python
extended_relationships = [
    ("Dominic Toretto", "Brian O'Conner"),
    ("Dominic Toretto", "Mia Toretto"),
    ("Dominic Toretto", "Letty Ortiz"),
    ("Brian O'Conner", "Mia Toretto"),
    ("Dominic Toretto", "Roman Pearce"),
    ("Dominic Toretto", "Tej Parker"),
    ("Dominic Toretto", "Han Lue"),
    ("Dominic Toretto", "Luke Hobbs"),
    ("Dominic Toretto", "Gisele Yashar"),
    ("Brian O'Conner", "Roman Pearce"),
    ("Brian O'Conner", "Agente Bilkins"),
    ("Roman Pearce", "Tej Parker"),
    ("Han Lue", "Gisele Yashar"),
    ("Luke Hobbs", "Deckard Shaw"),
    ("Dominic Toretto", "Jakob Toretto"),
    ("Dominic Toretto", "Elena Neves"),
    ("Mr. Nobody", "Dominic Toretto"),
    ("Mr. Nobody", "Luke Hobbs"),
    ("Mr. Nobody", "Ramsey"),
    ("Dominic Toretto", "Ramsey"),
    ("Tej Parker", "Ramsey"),
    ("Roman Pearce", "Ramsey"),
    ("Magdalene Shaw", "Deckard Shaw"),
    ("Magdalene Shaw", "Owen Shaw"),
    ("Magdalene Shaw", "Hattie Shaw"),
    ("Sean Boswell", "Han Lue"),
    ("Sean Boswell", "Neela"),
    ("Sean Boswell", "Twinkie"),
    ("Neela", "Twinkie"),
    ("Brian O'Conner", "Suki"),
    ("Roman Pearce", "Carter Verone"),
    ("Brian O'Conner", "Carter Verone"), 
    ("Brian O'Conner", "Monica"),
    ("Dominic Toretto", "Cypher"),
    ("Dominic Toretto", "Vince"),
    ("Dominic Toretto", "Leon"),
    ("Dominic Toretto", "Jesse"),
    ("Dominic Toretto", "Johnny Tran"),
    ("Dominic Toretto", "Fenix"),
    ("Dominic Toretto", "Reyes"),
    ("Dominic Toretto", "Dante"), 
    ("Dominic Toretto", "Owen Shaw"),
    ("Dominic Toretto", "Deckard Shaw"), 
    ("Brian O'Conner", "Johnny Tran"),
    ("Brian O'Conner", "Braga"),
    ("Brian O'Conner", "Reyes"), 
    ("Brian O'Conner", "Owen Shaw"), 
    ("Sean Boswell", "Takeshi"), 
    ("Dominic Toretto", "Rico Santos"),
    ("Dominic Toretto", "Tego Leo"),
    ("Reyes", "Dante"),
    ("Braga", "Fenix"),
    ("Deckard Shaw", "Owen Shaw"),
    ("Monica", "Carter Verone"),
    ("Letty Ortiz", "Owen Shaw"), 
    ("Jakob Toretto", "Elena Neves")
]
```

```python
G.add_edges_from(extended_relationships)
```

```python
pos = nx.spring_layout(G, k=15, iterations=50)
```

```python
plt.figure(figsize=(12, 8))
nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=2000, font_size=8, font_weight="bold", edge_color="gray")
plt.title("Grafo de personajes de Rápidos y Furiosos")
plt.show()
```

```python
shortest_path_lengths = dict(nx.all_pairs_shortest_path_length(G))
```

## -----------------------------Distancias--------------------------------

```python
for source, target_lengths in shortest_path_lengths.items():
    print(f"Distancias desde {source}:")
    for target, length in target_lengths.items():
        print(f"  hasta {target}: {length}")
```

## -----------------------------Excentricidad-----------------------------

```python
eccentricity = nx.eccentricity(G)
```

```python
print("\nExcentricidad de los nodos:")
for node, ecc in eccentricity.items():
    print(f"{node}: {ecc}")
```

## -----------------Radio-Diámetro-Centro del Grafo------------------

```python
radio = nx.radius(G)
```

```python
diameter = nx.diameter(G)
```

```python
center = nx.center(G)
```

```python
print(f"\nRadio del grafo: {radio}")
print(f"Diámetro del grafo: {diameter}")
print(f"Centro del grafo: {center}")
```
